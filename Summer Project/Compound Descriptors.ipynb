### Define Function

```python
def cid_to_smile(assay):
    cid = len(assay)
    smile = []
    for i in range(0, len(cid)):
        c = Compound.from_cid(cid[i])
    #     len(c)
    #     mf = c.molucular_formula
        smiles = c.canonical_smiles
        smile.append(smiles)
```

### All Descriptors

```python
for descriptor in Descriptors.descList:
  print(descriptor[0])
```

```python
def generate_descriptor_value(smiles, descriptor_function):
  mol = Chem.MolFromSmiles(smiles)
  value = None
  if mol is not None:
    value = descriptor_function(mol)
  return value
```

```python
df_all = zikainhibdf.copy()

for descriptor_tuple in Descriptors.descList:
  descriptor_name = descriptor_tuple[0]
  descriptor_function = descriptor_tuple[1]
  df_all[descriptor_name] = df_all['SMILES'].apply(lambda x: generate_descriptor_value(x, descriptor_function))

df_all
```

```python
# remove constant columns
df_all = df_all.loc[:,df_all.apply(pd.Series.nunique) != 1]

# replace all infinity values with nan
df_all = df_all.replace([np.inf, -np.inf], np.nan)

# remove columns with above .15 empty cells
thresh = len(df_all) * .15
df_all.dropna(thresh = thresh, axis = 1, inplace = True)

# replace all empty cells with mean of columns
column_means = df_all.mean()
df_all = df_all.fillna(column_means)

df_all
```

```python
df_all.to_csv('Compound All Descriptors.csv', index=False)
```
